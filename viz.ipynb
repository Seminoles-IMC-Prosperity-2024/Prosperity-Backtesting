```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import sympy

data0 = pd.read_csv('round-1-island-data-bottle/prices_round_1_day_0.csv', delimiter=';')
data1 = pd.read_csv('round-1-island-data-bottle/prices_round_1_day_-1.csv', delimiter=';')
data2 = pd.read_csv('round-1-island-data-bottle/prices_round_1_day_-2.csv', delimiter=';')

amethysts = pd.concat([data0[data0['product'] == 'AMETHYSTS'], data1[data1['product'] == 'AMETHYSTS'], data2[data2['product'] == 'AMETHYSTS'].reset_index()], axis=0).reset_index()
starfruit = pd.concat([data0[data0['product'] == 'STARFRUIT'], data1[data1['product'] == 'STARFRUIT'], data2[data2['product'] == 'STARFRUIT']], axis=0).reset_index()
```

```python
s_mean = starfruit['mid_price'].rolling(20).mean()
s_std = starfruit['mid_price'].rolling(20).std()

up = s_mean + 2 * s_std
low = s_mean - 2 * s_std
```

```python
s_mavg = starfruit['mid_price'].rolling(20).mean()
l_mavg = starfruit['mid_price'].rolling(50).mean()
```

```python
plt.plot(starfruit['mid_price'][:1000])
# plt.plot(up[:1000])
# plt.plot(low[:1000])

plt.plot(s_mavg[:1000], color='r')
plt.plot(l_mavg[:1000], color='b')
plt.show()
```
